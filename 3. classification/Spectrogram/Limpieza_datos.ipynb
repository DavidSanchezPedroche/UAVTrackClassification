##Limpieza de datos

En este programa se agrupa la limpieza de datos general propuesta en el análisis de atributos previo, con las técnicas que resultaron de más utilidad.
Se aplica sobre una carpeta con datos raw y se guardan los datos limpios en la carpeta de outputPath.

1. Borrar duplicados

2. Borrar columnas vacías

3. Borrar columnas con varianza 0

Al final, esta limpieza se aplicará sólo a las columnas con datos cinemáticos, para detectar posibles trayectorias que no aporten mucho.
Finalmente se seleccionan los siguientes atributos:

timestamp, x, y, z, vx, vy, vz, ax, ay, az , drone_model

```python
#IMPORTS

from cmath import nan
import math
import time
import pandas as pd
from os import listdir
from os.path import isfile, join
import numpy as np
import pandas as pd
import matplotlib.pyplot as mp

#PATHS
folder="Hexarotor"
mypath= r"/content/drive/MyDrive/TFG Celia/Código/Datos_raw/"+folder+"/DATA/"
outputPath = r"/content/drive/MyDrive/TFG Celia/Código/Datos_limpios/"+folder
```

```python
!unzip '/content/drive/MyDrive/TFG Celia/Código/Datos_raw/Hexarotor/hex2.zip' -d mypath
```

```python
from pathlib import Path

for f in listdir(mypath): 
  
  output = outputPath+'/'+str(f)
  my_file = Path(output)
  if isfile(join(mypath, f)) and not(my_file.exists()):
    instance = str(f)
    df = pd.DataFrame() #CLEAN DF
    df = pd.read_csv(mypath+instance,sep=',')
    print(instance)

    #DROP COLUMNS THAT HAVE 0 VALUES and variance = 0

    df.drop_duplicates()
    for col in df.columns:
        if len(df[col].value_counts()) == 0:
          df=df.drop(columns=col)
        elif df[col].var() == 0 and col!='drone_model' and col!='x' and col!='y' and col!='z': #if one dimension is = 0
          df=df.drop(columns=col)


    aggregation_functions = {'x': 'first','y': 'first','z': 'first','vx': 'first','vy': 'first','vz': 'first','ax': 'first','ay': 'first','az': 'first','drone_model':'first'}
    df_new = df.groupby(df['timestamp']).aggregate(aggregation_functions)
    df_new.dropna(inplace=True)
    print("Instance shape: "+str(df_new.shape))
    df_new.to_csv(outputPath+'/'+instance)   
```
